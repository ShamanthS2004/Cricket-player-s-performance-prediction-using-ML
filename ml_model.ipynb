```python
import pandas as pd
import tkinter as tk
from tkinter import ttk
import numpy as np
from PIL import Image, ImageTk

# Load CSV files
path = r'E:\Major project\\'

# Existing teams
ind_batting_vs_aus = pd.read_csv(path + 'ind_batting_against_aus.csv')
ind_bowling_vs_aus = pd.read_csv(path + 'ind_bowling_against_aus.csv')
aus_batting_vs_ind = pd.read_csv(path + 'aus_batting_against_ind.csv')
aus_bowling_vs_ind = pd.read_csv(path + 'aus_bowling_against_ind.csv')

ind_batting_vs_eng = pd.read_csv(path + 'ind_batting_against_eng.csv')
ind_bowling_vs_eng = pd.read_csv(path + 'ind_bowling_against_eng.csv')
eng_batting_vs_ind = pd.read_csv(path + 'eng_batting_against_ind.csv')
eng_bowling_vs_ind = pd.read_csv(path + 'eng_bowling_against_ind.csv')

ind_batting_vs_pak = pd.read_csv(path + 'ind_batting_against_pak.csv')
ind_bowling_vs_pak = pd.read_csv(path + 'ind_bowling_against_pak.csv')
pak_batting_vs_ind = pd.read_csv(path + 'pak_batting_against_ind.csv')
pak_bowling_vs_ind = pd.read_csv(path + 'pak_bowling_against_ind.csv')

ind_batting_vs_sa = pd.read_csv(path + 'ind_batting_against_sa.csv')
ind_bowling_vs_sa = pd.read_csv(path + 'ind_bowling_against_sa.csv')
sa_batting_vs_ind = pd.read_csv(path + 'sa_batting_against_ind.csv')
sa_bowling_vs_ind = pd.read_csv(path + 'sa_bowling_against_ind.csv')

ind_batting_vs_sri = pd.read_csv(path + 'ind_batting_against_sri.csv')
ind_bowling_vs_sri = pd.read_csv(path + 'ind_bowling_against_sri.csv')
sri_batting_vs_ind = pd.read_csv(path + 'sri_batting_against_ind.csv')
sri_bowling_vs_ind = pd.read_csv(path + 'sri_bowling_against_ind.csv')

ind_batting_vs_afg = pd.read_csv(path + 'ind_batting_against_afg.csv')
ind_bowling_vs_afg = pd.read_csv(path + 'ind_bowling_against_afg.csv')
afg_batting_vs_ind = pd.read_csv(path + 'afg_batting_against_ind.csv')
afg_bowling_vs_ind = pd.read_csv(path + 'afg_bowling_against_ind.csv')

ind_batting_vs_nz = pd.read_csv(path + 'ind_batting_against_nz.csv')
ind_bowling_vs_nz = pd.read_csv(path + 'ind_bowling_against_nz.csv')
nz_batting_vs_ind = pd.read_csv(path + 'nz_batting_against_ind.csv')
nz_bowling_vs_ind = pd.read_csv(path + 'nz_bowling_against_ind.csv')

# Additional teams
ind_batting_vs_wi = pd.read_csv(path + 'ind_batting_against_wi.csv')
ind_bowling_vs_wi = pd.read_csv(path + 'ind_bowling_against_wi.csv')
wi_batting_vs_ind = pd.read_csv(path + 'wi_batting_against_ind.csv')
wi_bowling_vs_ind = pd.read_csv(path + 'wi_bowling_against_ind.csv')

ind_batting_vs_ban = pd.read_csv(path + 'ind_batting_against_ban.csv')
ind_bowling_vs_ban = pd.read_csv(path + 'ind_bowling_against_ban.csv')
ban_batting_vs_ind = pd.read_csv(path + 'ban_batting_against_ind.csv')
ban_bowling_vs_ind = pd.read_csv(path + 'ban_bowling_against_ind.csv')

#Australia team 
#Eng
aus_batting_vs_eng = pd.read_csv(path + 'aus_batting_against_eng.csv')
aus_bowling_vs_eng = pd.read_csv(path + 'aus_bowling_against_eng.csv')
eng_batting_vs_aus = pd.read_csv(path + 'eng_batting_against_aus.csv')
eng_bowling_vs_aus = pd.read_csv(path + 'eng_bowling_against_aus.csv')
#Pak
aus_batting_vs_pak = pd.read_csv(path + 'aus_batting_against_pak.csv')
aus_bowling_vs_pak = pd.read_csv(path + 'aus_bowling_against_pak.csv')
pak_batting_vs_aus = pd.read_csv(path + 'pak_batting_against_aus.csv')
pak_bowling_vs_aus = pd.read_csv(path + 'pak_bowling_against_aus.csv')
#SA
aus_batting_vs_sa = pd.read_csv(path + 'aus_batting_against_sa.csv')
aus_bowling_vs_sa = pd.read_csv(path + 'aus_bowling_against_sa.csv')
sa_batting_vs_aus = pd.read_csv(path + 'sa_batting_against_aus.csv')
sa_bowling_vs_aus = pd.read_csv(path + 'sa_bowling_against_aus.csv')
#Sri
aus_batting_vs_sri = pd.read_csv(path + 'aus_batting_against_sri.csv')
aus_bowling_vs_sri = pd.read_csv(path + 'aus_bowling_against_sri.csv')
sri_batting_vs_aus = pd.read_csv(path + 'sri_batting_against_aus.csv')
sri_bowling_vs_aus = pd.read_csv(path + 'sri_bowling_against_aus.csv')
#Afg
aus_batting_vs_afg = pd.read_csv(path + 'aus_batting_against_afg.csv')
aus_bowling_vs_afg = pd.read_csv(path + 'aus_bowling_against_afg.csv')
afg_batting_vs_aus = pd.read_csv(path + 'afg_batting_against_aus.csv')
afg_bowling_vs_aus = pd.read_csv(path + 'afg_bowling_against_aus.csv')
#NZ
aus_batting_vs_nz = pd.read_csv(path + 'aus_batting_against_nz.csv')
aus_bowling_vs_nz = pd.read_csv(path + 'aus_bowling_against_nz.csv')
nz_batting_vs_aus = pd.read_csv(path + 'nz_batting_against_aus.csv')
nz_bowling_vs_aus = pd.read_csv(path + 'nz_bowling_against_aus.csv')
#Wi
aus_batting_vs_wi = pd.read_csv(path + 'aus_batting_against_wi.csv')
aus_bowling_vs_wi = pd.read_csv(path + 'aus_bowling_against_wi.csv')
wi_batting_vs_aus = pd.read_csv(path + 'wi_batting_against_aus.csv')
wi_bowling_vs_aus = pd.read_csv(path + 'wi_bowling_against_aus.csv')
#Ban
aus_batting_vs_ban = pd.read_csv(path + 'aus_batting_against_ban.csv')
aus_bowling_vs_ban = pd.read_csv(path + 'aus_bowling_against_ban.csv')
ban_batting_vs_aus = pd.read_csv(path + 'ban_batting_against_aus.csv')
ban_bowling_vs_aus = pd.read_csv(path + 'ban_bowling_against_aus.csv')

#England against teams! 
#Pakistan
eng_batting_vs_pak = pd.read_csv(path + 'eng_batting_against_pak.csv')
eng_bowling_vs_pak = pd.read_csv(path + 'eng_bowling_against_pak.csv')
pak_batting_vs_eng = pd.read_csv(path + 'pak_batting_against_eng.csv')
pak_bowling_vs_eng = pd.read_csv(path + 'pak_bowling_against_eng.csv')
#South Africa
eng_batting_vs_sa = pd.read_csv(path + 'eng_batting_against_sa.csv')
eng_bowling_vs_sa = pd.read_csv(path + 'eng_bowling_against_sa.csv')
sa_batting_vs_eng = pd.read_csv(path + 'sa_batting_against_eng.csv')
sa_bowling_vs_eng = pd.read_csv(path + 'sa_bowling_against_eng.csv')
#Sri Lanka
eng_batting_vs_sri = pd.read_csv(path + 'eng_batting_against_sri.csv')
eng_bowling_vs_sri = pd.read_csv(path + 'eng_bowling_against_sri.csv')
sri_batting_vs_eng = pd.read_csv(path + 'sri_batting_against_eng.csv')
sri_bowling_vs_eng = pd.read_csv(path + 'sri_bowling_against_eng.csv')
#Afghanistan
eng_batting_vs_afg = pd.read_csv(path + 'eng_batting_against_afg.csv')
eng_bowling_vs_afg = pd.read_csv(path + 'eng_bowling_against_afg.csv')
afg_batting_vs_eng = pd.read_csv(path + 'afg_batting_against_eng.csv')
afg_bowling_vs_eng = pd.read_csv(path + 'afg_bowling_against_eng.csv')
#New Zealand
eng_batting_vs_nz = pd.read_csv(path + 'eng_batting_against_nz.csv')
eng_bowling_vs_nz = pd.read_csv(path + 'eng_bowling_against_nz.csv')
nz_batting_vs_eng = pd.read_csv(path + 'nz_batting_against_eng.csv')
nz_bowling_vs_eng = pd.read_csv(path + 'nz_bowling_against_eng.csv')
#West Indies
eng_batting_vs_wi = pd.read_csv(path + 'eng_batting_against_wi.csv')
eng_bowling_vs_wi = pd.read_csv(path + 'eng_bowling_against_wi.csv')
wi_batting_vs_eng = pd.read_csv(path + 'wi_batting_against_eng.csv')
wi_bowling_vs_eng = pd.read_csv(path + 'wi_bowling_against_eng.csv')
#Bangladesh
eng_batting_vs_ban = pd.read_csv(path + 'eng_batting_against_ban.csv')
eng_bowling_vs_ban = pd.read_csv(path + 'eng_bowling_against_ban.csv')
ban_batting_vs_eng = pd.read_csv(path + 'ban_batting_against_eng.csv')
ban_bowling_vs_eng = pd.read_csv(path + 'ban_bowling_against_eng.csv')
#Pakistan vs other teams
#South Africa
pak_batting_vs_sa = pd.read_csv(path + 'pak_batting_against_sa.csv')
pak_bowling_vs_sa = pd.read_csv(path + 'pak_bowling_against_sa.csv')
sa_batting_vs_pak = pd.read_csv(path + 'sa_batting_against_pak.csv')
sa_bowling_vs_pak = pd.read_csv(path + 'sa_bowling_against_pak.csv')
#Sri Lanka
pak_batting_vs_sri = pd.read_csv(path + 'pak_batting_against_sri.csv')
pak_bowling_vs_sri = pd.read_csv(path + 'pak_bowling_against_sri.csv')
sri_batting_vs_pak = pd.read_csv(path + 'sri_batting_against_pak.csv')
sri_bowling_vs_pak = pd.read_csv(path + 'sri_bowling_against_pak.csv')
#Afghanistan
pak_batting_vs_afg = pd.read_csv(path + 'pak_batting_against_afg.csv')
pak_bowling_vs_afg = pd.read_csv(path + 'pak_bowling_against_afg.csv')
afg_batting_vs_pak = pd.read_csv(path + 'afg_batting_against_pak.csv')
afg_bowling_vs_pak = pd.read_csv(path + 'afg_bowling_against_pak.csv')
#New Zealand
pak_batting_vs_nz = pd.read_csv(path + 'pak_batting_against_nz.csv')
pak_bowling_vs_nz = pd.read_csv(path + 'pak_bowling_against_nz.csv')
nz_batting_vs_pak = pd.read_csv(path + 'nz_batting_against_pak.csv')
nz_bowling_vs_pak = pd.read_csv(path + 'nz_bowling_against_pak.csv')
#West Indies
pak_batting_vs_wi = pd.read_csv(path + 'pak_batting_against_wi.csv')
pak_bowling_vs_wi = pd.read_csv(path + 'pak_bowling_against_wi.csv')
wi_batting_vs_pak = pd.read_csv(path + 'wi_batting_against_pak.csv')
wi_bowling_vs_pak = pd.read_csv(path + 'wi_bowling_against_pak.csv')
#Bangladesh
pak_batting_vs_ban = pd.read_csv(path + 'pak_batting_against_ban.csv')
pak_bowling_vs_ban = pd.read_csv(path + 'pak_bowling_against_ban.csv')
ban_batting_vs_pak = pd.read_csv(path + 'ban_batting_against_pak.csv')
ban_bowling_vs_pak = pd.read_csv(path + 'ban_bowling_against_pak.csv')
#South Africa vs all teams!
#Afghanistan
sa_batting_vs_afg = pd.read_csv(path + 'sa_batting_against_afg.csv')
sa_bowling_vs_afg = pd.read_csv(path + 'sa_bowling_against_afg.csv')
afg_batting_vs_sa = pd.read_csv(path + 'afg_batting_against_sa.csv')
afg_bowling_vs_sa = pd.read_csv(path + 'afg_bowling_against_sa.csv')
#New Zealand
sa_batting_vs_nz = pd.read_csv(path + 'sa_batting_against_nz.csv')
sa_bowling_vs_nz = pd.read_csv(path + 'sa_bowling_against_nz.csv')
nz_batting_vs_sa = pd.read_csv(path + 'nz_batting_against_sa.csv')
nz_bowling_vs_sa = pd.read_csv(path + 'nz_bowling_against_sa.csv')
#West Indies
sa_batting_vs_wi = pd.read_csv(path + 'sa_batting_against_wi.csv')
sa_bowling_vs_wi = pd.read_csv(path + 'sa_bowling_against_wi.csv')
wi_batting_vs_sa = pd.read_csv(path + 'wi_batting_against_sa.csv')
wi_bowling_vs_sa = pd.read_csv(path + 'wi_bowling_against_sa.csv')
#Bangladesh
sa_batting_vs_ban = pd.read_csv(path + 'sa_batting_against_ban.csv')
sa_bowling_vs_ban = pd.read_csv(path + 'sa_bowling_against_ban.csv')
ban_batting_vs_sa = pd.read_csv(path + 'ban_batting_against_sa.csv')
ban_bowling_vs_sa = pd.read_csv(path + 'ban_bowling_against_sa.csv')
#Sri Lanka vs all teams!
#Afghanistan
sri_batting_vs_afg = pd.read_csv(path + 'sri_batting_against_afg.csv')
sri_bowling_vs_afg = pd.read_csv(path + 'sri_bowling_against_afg.csv')
afg_batting_vs_sri = pd.read_csv(path + 'afg_batting_against_sri.csv')
afg_bowling_vs_sri = pd.read_csv(path + 'afg_bowling_against_sri.csv')
#New Zealand
sri_batting_vs_nz = pd.read_csv(path + 'sri_batting_against_nz.csv')
sri_bowling_vs_nz = pd.read_csv(path + 'sri_bowling_against_nz.csv')
nz_batting_vs_sri = pd.read_csv(path + 'nz_batting_against_sri.csv')
nz_bowling_vs_sri = pd.read_csv(path + 'nz_bowling_against_sri.csv')
#Bangladesh
sri_batting_vs_ban = pd.read_csv(path + 'sri_batting_against_ban.csv')
sri_bowling_vs_ban = pd.read_csv(path + 'sri_bowling_against_ban.csv')
ban_batting_vs_sri = pd.read_csv(path + 'ban_batting_against_sri.csv')
ban_bowling_vs_sri = pd.read_csv(path + 'ban_bowling_against_sri.csv')
#Afghanistan vs all teams!
#West Indies
afg_batting_vs_wi = pd.read_csv(path + 'afg_batting_against_wi.csv')
afg_bowling_vs_wi = pd.read_csv(path + 'afg_bowling_against_wi.csv')
wi_batting_vs_afg = pd.read_csv(path + 'wi_batting_against_afg.csv')
wi_bowling_vs_afg = pd.read_csv(path + 'wi_bowling_against_afg.csv')
#Bangladesh
afg_batting_vs_ban = pd.read_csv(path + 'afg_batting_against_ban.csv')
afg_bowling_vs_ban = pd.read_csv(path + 'afg_bowling_against_ban.csv')
ban_batting_vs_afg = pd.read_csv(path + 'ban_batting_against_afg.csv')
ban_bowling_vs_afg = pd.read_csv(path + 'ban_bowling_against_afg.csv')
#New zealand vs all teams!
#West Indies
nz_batting_vs_wi = pd.read_csv(path + 'nz_batting_against_wi.csv')
nz_bowling_vs_wi = pd.read_csv(path + 'nz_bowling_against_wi.csv')
wi_batting_vs_nz = pd.read_csv(path + 'wi_batting_against_nz.csv')
wi_bowling_vs_nz = pd.read_csv(path + 'wi_bowling_against_nz.csv')
#Bangladesh
nz_batting_vs_ban = pd.read_csv(path + 'nz_batting_against_ban.csv')
nz_bowling_vs_ban = pd.read_csv(path + 'nz_bowling_against_ban.csv')
ban_batting_vs_nz = pd.read_csv(path + 'ban_batting_against_nz.csv')
ban_bowling_vs_nz = pd.read_csv(path + 'ban_bowling_against_nz.csv')
#West Indies vs all teams!
wi_batting_vs_ban = pd.read_csv(path + 'wi_batting_against_ban.csv')
wi_bowling_vs_ban = pd.read_csv(path + 'wi_bowling_against_ban.csv')
ban_batting_vs_wi = pd.read_csv(path + 'ban_batting_against_wi.csv')
ban_bowling_vs_wi = pd.read_csv(path + 'ban_bowling_against_wi.csv')

#West Indies
sri_batting_vs_wi = pd.read_csv(path + 'sri_batting_against_wi.csv')
sri_bowling_vs_wi = pd.read_csv(path + 'sri_bowling_against_wi.csv')
wi_batting_vs_sri = pd.read_csv(path + 'wi_batting_against_sri.csv')
wi_bowling_vs_sri = pd.read_csv(path + 'wi_bowling_against_sri.csv')

stadium_conditions = pd.read_csv(path + 'stadium_conditions.csv')

# Rename stadium conditions columns
stadium_conditions.rename(columns={'Stadium Name': 'stadium_name'}, inplace=True)

# Check if 'Condition' column exists, add if missing
if 'Condition' not in stadium_conditions.columns:
    print("Warning: Column 'Condition' is missing from stadium_conditions DataFrame. Adding default values.")
    stadium_conditions['Condition'] = 'Neutral'  # Default value

# Renaming dictionaries
batting_rename_dict = {
    'Avg': 'avg',
    'Strike Rate': 'strike_rate',
    'Runs': 'runs',
    'Strike rate': 'strike_rate',
    'Not out': 'Not Out'
}

bowling_rename_dict = {
    'Avg': 'avg',
    'Ave': 'avg',
    'Econ': 'econ'
}

# Rename columns to ensure consistency
def rename_columns(df, rename_dict):
    df.rename(columns=rename_dict, inplace=True)
    return df

# Apply renaming
dataframes = [
    ind_batting_vs_aus, ind_bowling_vs_aus, aus_batting_vs_ind, aus_bowling_vs_ind,
    ind_batting_vs_eng, ind_bowling_vs_eng, eng_batting_vs_ind, eng_bowling_vs_ind,
    ind_batting_vs_pak, ind_bowling_vs_pak, pak_batting_vs_ind, pak_bowling_vs_ind,
    ind_batting_vs_sa, ind_bowling_vs_sa, sa_batting_vs_ind, sa_bowling_vs_ind,
    ind_batting_vs_sri, ind_bowling_vs_sri, sri_batting_vs_ind, sri_bowling_vs_ind,
    ind_batting_vs_afg, ind_bowling_vs_afg, afg_batting_vs_ind, afg_bowling_vs_ind,
    ind_batting_vs_nz, ind_bowling_vs_nz, nz_batting_vs_ind, nz_bowling_vs_ind,
    ind_batting_vs_wi, ind_bowling_vs_wi, wi_batting_vs_ind, wi_bowling_vs_ind,
    ind_batting_vs_ban, ind_bowling_vs_ban, ban_batting_vs_ind, ban_bowling_vs_ind,
    aus_batting_vs_eng, aus_bowling_vs_eng ,eng_batting_vs_aus, eng_bowling_vs_aus,
    aus_batting_vs_pak,aus_bowling_vs_pak, pak_batting_vs_aus,pak_bowling_vs_aus ,
    aus_batting_vs_sa ,aus_bowling_vs_sa,sa_batting_vs_aus,sa_bowling_vs_aus,
    aus_batting_vs_sri,aus_bowling_vs_sri,sri_batting_vs_aus,sri_bowling_vs_aus,
    aus_batting_vs_afg ,aus_bowling_vs_afg ,afg_batting_vs_aus,afg_bowling_vs_aus,
    aus_batting_vs_nz , aus_bowling_vs_nz ,nz_batting_vs_aus,nz_bowling_vs_aus,
    aus_batting_vs_wi , aus_bowling_vs_wi,wi_batting_vs_aus,wi_bowling_vs_aus,
    aus_batting_vs_ban, aus_bowling_vs_ban,ban_batting_vs_aus ,ban_bowling_vs_aus,
    eng_batting_vs_pak, eng_bowling_vs_pak,pak_batting_vs_eng,pak_bowling_vs_eng,
    eng_batting_vs_sa,eng_bowling_vs_sa ,sa_batting_vs_eng, sa_bowling_vs_eng,
    eng_batting_vs_sri ,eng_bowling_vs_sri ,sri_batting_vs_eng ,sri_bowling_vs_eng,
    eng_batting_vs_afg,eng_bowling_vs_afg ,eng_bowling_vs_afg ,afg_bowling_vs_eng,
    eng_batting_vs_nz,eng_bowling_vs_nz,nz_batting_vs_eng ,nz_bowling_vs_eng,
    eng_batting_vs_wi,eng_bowling_vs_wi , wi_batting_vs_eng,wi_bowling_vs_eng,
    eng_batting_vs_ban,eng_bowling_vs_ban , ban_batting_vs_eng ,ban_bowling_vs_eng,
    
    pak_batting_vs_sri , pak_bowling_vs_sri,sri_batting_vs_pak,sri_bowling_vs_pak,
    pak_batting_vs_afg, pak_bowling_vs_afg,afg_batting_vs_pak,afg_bowling_vs_pak,
    pak_batting_vs_nz,pak_bowling_vs_nz,nz_batting_vs_pak,nz_bowling_vs_pak,
    pak_batting_vs_wi,pak_bowling_vs_wi ,wi_batting_vs_pak,wi_bowling_vs_pak,
    pak_batting_vs_ban ,pak_bowling_vs_ban,ban_batting_vs_pak,ban_bowling_vs_pak,
    pak_batting_vs_sa, pak_bowling_vs_sa , sa_batting_vs_pak,sa_bowling_vs_pak ,


    sa_batting_vs_afg,sa_bowling_vs_afg ,afg_batting_vs_sa,afg_bowling_vs_sa,
    sa_batting_vs_nz ,sa_bowling_vs_nz,nz_batting_vs_sa,nz_bowling_vs_sa,
    sa_batting_vs_wi,sa_bowling_vs_wi,wi_batting_vs_sa ,wi_bowling_vs_sa,
    sa_batting_vs_ban,sa_bowling_vs_ban,ban_batting_vs_sa,ban_bowling_vs_sa,

    sri_batting_vs_afg,sri_bowling_vs_afg ,afg_batting_vs_sri,afg_bowling_vs_sri,
    sri_batting_vs_nz,sri_bowling_vs_nz,nz_batting_vs_sri,nz_bowling_vs_sri,
    sri_batting_vs_ban,sri_bowling_vs_ban, ban_batting_vs_sri,ban_bowling_vs_sri,

    afg_batting_vs_wi,afg_bowling_vs_wi,wi_batting_vs_afg,wi_bowling_vs_afg,
    afg_batting_vs_ban,afg_bowling_vs_ban,ban_batting_vs_afg,ban_bowling_vs_afg,

    nz_batting_vs_wi,nz_bowling_vs_wi,wi_batting_vs_nz,wi_bowling_vs_nz,
    nz_batting_vs_ban,nz_bowling_vs_ban,ban_batting_vs_nz,ban_bowling_vs_nz,

    wi_batting_vs_ban,wi_bowling_vs_ban,ban_batting_vs_wi ,ban_bowling_vs_wi,
    sri_batting_vs_wi,sri_bowling_vs_wi,wi_batting_vs_sri,wi_bowling_vs_sri
    

    
]

for df in dataframes:
    rename_columns(df, batting_rename_dict)
    rename_columns(df, bowling_rename_dict)

# Convert relevant columns to numeric
def convert_columns_to_numeric(df, columns):
    for column in columns:
        if column in df.columns:
            df[column] = pd.to_numeric(df[column], errors='coerce')
        else:
            print(f"Warning: Column '{column}' not found in DataFrame")
    return df

required_batting_columns = ['avg', 'strike_rate', 'runs']
required_bowling_columns = ['avg', 'econ']

# Apply conversion
for df in dataframes:
    convert_columns_to_numeric(df, required_batting_columns)
    convert_columns_to_numeric(df, required_bowling_columns)

# Fill NaN values with relevant random values
def fill_na_with_random(df, columns):
    for column in columns:
        if column in df.columns:
            mean_value = df[column].mean()
            std_value = df[column].std()
            random_values = np.random.normal(mean_value, std_value, df[column].isna().sum())
            df.loc[df[column].isna(), column] = random_values
        else:
            print(f"Warning: Column '{column}' not found in DataFrame")
    return df

# Apply fill_na_with_random
for df in dataframes:
    fill_na_with_random(df, required_batting_columns)
    fill_na_with_random(df, required_bowling_columns)

# Ensure columns exist in the DataFrame
def ensure_columns_exist(df, columns):
    for column in columns:
        if column not in df.columns:
            if column == 'econ':
                df[column] = np.random.normal(4.0, 0.5, len(df))  # Assuming an average economy of 4.0
            else:
                df[column] = np.nan
    return df

# Apply ensure_columns_exist
for df in dataframes:
    ensure_columns_exist(df, required_batting_columns)
    ensure_columns_exist(df, required_bowling_columns)


# Select best 11 players with a balance of 6 batsmen and 5 bowlers, adjusted for stadium conditions
def select_best_11(batting_df, bowling_df, stadium_info):
    # Ensure required columns exist
    batting_df = ensure_columns_exist(batting_df, ['avg', 'strike_rate'])
    bowling_df = ensure_columns_exist(bowling_df, ['avg', 'econ'])

    # Determine stadium preference
    if stadium_info['Condition'].values[0] == 'Batting':
        batting_weight = 1.5
        bowling_weight = 0.75
    elif stadium_info['Condition'].values[0] == 'Bowling':
        batting_weight = 0.75
        bowling_weight = 1.5
    else:
        batting_weight = 1.0
        bowling_weight = 1.0

    # Adjust player metrics based on stadium conditions and add randomness
    batting_df['adjusted_avg'] = batting_df['avg'] * batting_weight * np.random.uniform(0.9, 1.1, len(batting_df))
    bowling_df['adjusted_avg'] = bowling_df['avg'] * bowling_weight * np.random.uniform(0.9, 1.1, len(bowling_df))

    # Select top 6 batsmen based on adjusted average
    best_batsmen = batting_df.sort_values(by='adjusted_avg', ascending=False).head(6)
    # Select top 5 bowlers based on adjusted average
    best_bowlers = bowling_df.sort_values(by='adjusted_avg', ascending=True).head(5)

    # Combine all selected players
    best_11 = pd.concat([best_batsmen, best_bowlers], ignore_index=True).drop_duplicates(subset='Player')

    # Ensure exactly 11 players
    if len(best_11) < 11:
        # If less than 11 players are found, add more from remaining players
        remaining_players = pd.concat([batting_df, bowling_df]).drop_duplicates(subset='Player')
        additional_players = remaining_players[~remaining_players['Player'].isin(best_11['Player'])].head(11 - len(best_11))
        best_11 = pd.concat([best_11, additional_players], ignore_index=True).drop_duplicates(subset='Player')

    return best_11

# Function to calculate winning probability
def calculate_winning_probability(team_batting_data, opponent_batting_data, team_bowling_data, opponent_bowling_data):
    # Basic averages and strike rates for team and opponent
    team_batting_avg = team_batting_data['avg'].mean()
    opponent_batting_avg = opponent_batting_data['avg'].mean()
    team_bowling_avg = team_bowling_data['avg'].mean()
    opponent_bowling_avg = opponent_bowling_data['avg'].mean()
    
    # Simple probability calculation based on batting and bowling averages
    batting_strength = team_batting_avg / (team_batting_avg + opponent_batting_avg)
    bowling_strength = opponent_bowling_avg / (team_bowling_avg + opponent_bowling_avg)
    
    # Combined winning probability (simplified)
    winning_probability = (batting_strength + (1 - bowling_strength)) / 2
    
    return winning_probability

def adjust_metrics_for_stadium(df, stadium_name, player_type):
    """Adjust player metrics based on stadium conditions and player type."""
    stadium_info = stadium_conditions[stadium_conditions['stadium_name'] == stadium_name]

    if stadium_info.empty:
        print(f"Warning: Stadium '{stadium_name}' not found.")
        return df  # Return original DataFrame if stadium not found

    condition = stadium_info['Condition'].values[0]

    # Adjust metrics based on stadium condition
    if condition in ["Batting Friendly", "Small Stadium"]:
        if player_type in ["Batsman", "All-Rounder"]:
            df['avg'] *= 1.05  # Smaller adjustment (5% increase)
            df['strike_rate'] *= 1.05  # Smaller adjustment (5% increase)
        if player_type in ["All-Rounder", "Bowler"]:
            df['econ'] *= 1.1  # Increase economy slightly (worse performance)
            df['strike_rate'] *= 1.1  # Increase bowling strike rate (worse performance)
    elif condition in ["Bowling Friendly", "Large Stadium"]:
        if player_type in ["Batsman", "All-Rounder"]:
            df['avg'] *= 0.95  # Slight decrease in average (5% decrease)
            df['strike_rate'] *= 0.95  # Slight decrease in strike rate (5% decrease)
        if player_type in ["All-Rounder", "Bowler"]:
            df['econ'] *= 0.9  # Lower economy = better performance (10% improvement)
            df['strike_rate'] *= 0.9  # Lower bowling strike rate = better performance

    return df


# GUI Implementation
class CricketTeamSelector:
    def __init__(self, root):
        self.root = root
        self.root.title("Cricket Team Selector")
        self.root.geometry("600x500")

        self.team_label = tk.Label(root, text="Select Team:")
        self.team_label.pack()

        self.team_combobox = ttk.Combobox(root, values=["India", "Australia", "England", "Pakistan", "South Africa", "Sri Lanka", "Afghanistan", "New Zealand", "West Indies", "Bangladesh"])
        self.team_combobox.pack()

        self.opponent_label = tk.Label(root, text="Select Opponent Team:")
        self.opponent_label.pack()

        # Ensure "India" is included in the opponent team list
        self.opponent_combobox = ttk.Combobox(root, values=["India", "Australia", "England", "Pakistan", "South Africa", "Sri Lanka", "Afghanistan", "New Zealand", "West Indies", "Bangladesh"])
        self.opponent_combobox.pack()

        self.stadium_label = tk.Label(root, text="Select Stadium:")
        self.stadium_label.pack()

        self.stadium_combobox = ttk.Combobox(root, values=stadium_conditions['stadium_name'].tolist())
        self.stadium_combobox.pack()

        self.submit_button = tk.Button(root, text="Show Best 11 and Win Probability", command=self.show_team)
        self.submit_button.pack()

        self.team_output = tk.Text(root, height=20, width=80)
        self.team_output.pack()

    def show_team(self):
        selected_team = self.team_combobox.get()
        opponent_team = self.opponent_combobox.get()
        stadium_name = self.stadium_combobox.get()

        if stadium_name in stadium_conditions['stadium_name'].values:
            stadium_info = stadium_conditions[stadium_conditions['stadium_name'] == stadium_name]
        else:
            self.team_output.delete(1.0, tk.END)
            self.team_output.insert(tk.END, "Invalid stadium selection.")
            return

        # Determine which DataFrame to use based on selected teams
        team_batting_data = None
        team_bowling_data = None
        opponent_batting_data = None
        opponent_bowling_data = None

        if selected_team == "India" and opponent_team == "Australia":
            team_batting_data = ind_batting_vs_aus
            team_bowling_data = ind_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_ind
            opponent_bowling_data = aus_bowling_vs_ind
        elif selected_team == "Australia" and opponent_team == "India":
            team_batting_data = aus_batting_vs_ind
            team_bowling_data = aus_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_aus
            opponent_bowling_data = ind_bowling_vs_aus
        elif selected_team == "India" and opponent_team == "England":
            team_batting_data = ind_batting_vs_eng
            team_bowling_data = ind_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_ind
            opponent_bowling_data = eng_bowling_vs_ind
        elif selected_team == "England" and opponent_team == "India":
            team_batting_data = eng_batting_vs_ind
            team_bowling_data = eng_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_eng
            opponent_bowling_data = ind_bowling_vs_eng
        elif selected_team == "India" and opponent_team == "Pakistan":
            team_batting_data = ind_batting_vs_pak
            team_bowling_data = ind_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_ind
            opponent_bowling_data = pak_bowling_vs_ind
        elif selected_team == "Pakistan" and opponent_team == "India":
            team_batting_data = pak_batting_vs_ind
            team_bowling_data = pak_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_pak
            opponent_bowling_data = ind_bowling_vs_pak
        elif selected_team == "India" and opponent_team == "South Africa":
            team_batting_data = ind_batting_vs_sa
            team_bowling_data = ind_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_ind
            opponent_bowling_data = sa_bowling_vs_ind
        elif selected_team == "South Africa" and opponent_team == "India":
            team_batting_data = sa_batting_vs_ind
            team_bowling_data = sa_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_sa
            opponent_bowling_data = ind_bowling_vs_sa
        elif selected_team == "India" and opponent_team == "Sri Lanka":
            team_batting_data = ind_batting_vs_sri
            team_bowling_data = ind_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_ind
            opponent_bowling_data = sri_bowling_vs_ind
        elif selected_team == "Sri Lanka" and opponent_team == "India":
            team_batting_data = sri_batting_vs_ind
            team_bowling_data = sri_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_sri
            opponent_bowling_data = ind_bowling_vs_sri
        elif selected_team == "India" and opponent_team == "Afghanistan":
            team_batting_data = ind_batting_vs_afg
            team_bowling_data = ind_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_ind
            opponent_bowling_data = afg_bowling_vs_ind
        elif selected_team == "Afghanistan" and opponent_team == "India":
            team_batting_data = afg_batting_vs_ind
            team_bowling_data = afg_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_afg
            opponent_bowling_data = ind_bowling_vs_afg
        elif selected_team == "India" and opponent_team == "New Zealand":
            team_batting_data = ind_batting_vs_nz
            team_bowling_data = ind_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_ind
            opponent_bowling_data = nz_bowling_vs_ind
        elif selected_team == "New Zealand" and opponent_team == "India":
            team_batting_data = nz_batting_vs_ind
            team_bowling_data = nz_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_nz
            opponent_bowling_data = ind_bowling_vs_nz
        elif selected_team == "India" and opponent_team == "West Indies":
            team_batting_data = ind_batting_vs_wi
            team_bowling_data = ind_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_ind
            opponent_bowling_data = wi_bowling_vs_ind
        elif selected_team == "West Indies" and opponent_team == "India":
            team_batting_data = wi_batting_vs_ind
            team_bowling_data = wi_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_wi
            opponent_bowling_data = ind_bowling_vs_wi
        elif selected_team == "India" and opponent_team == "Bangladesh":
            team_batting_data = ind_batting_vs_ban
            team_bowling_data = ind_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_ind
            opponent_bowling_data = ban_bowling_vs_ind
        elif selected_team == "Bangladesh" and opponent_team == "India":
            team_batting_data = ban_batting_vs_ind
            team_bowling_data = ban_bowling_vs_ind
            opponent_batting_data = ind_batting_vs_ban
            opponent_bowling_data = ind_bowling_vs_ban
        elif selected_team == "Australia" and opponent_team == "England":
            team_batting_data = aus_batting_vs_eng
            team_bowling_data = aus_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_aus
            opponent_bowling_data = eng_bowling_vs_aus
        elif selected_team == "England" and opponent_team == "Australia" :
            team_batting_data = eng_batting_vs_aus
            team_bowling_data = eng_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_eng
            opponent_bowling_data = aus_bowling_vs_eng    
        elif selected_team == "Australia" and opponent_team == "Pakistan":
            team_batting_data = aus_batting_vs_pak
            team_bowling_data = aus_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_aus
            opponent_bowling_data = pak_bowling_vs_aus
        elif selected_team == "Pakistan" and opponent_team == "Australia" :
            team_batting_data = pak_batting_vs_aus
            team_bowling_data = pak_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_pak
            opponent_bowling_data = aus_bowling_vs_pak        
        elif selected_team == "Australia" and opponent_team == "South Africa":
            team_batting_data = aus_batting_vs_sa
            team_bowling_data = aus_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_aus
            opponent_bowling_data = sa_bowling_vs_aus
        elif selected_team == "South Africa" and opponent_team == "Australia" :
            team_batting_data = sa_batting_vs_aus
            team_bowling_data = sa_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_sa
            opponent_bowling_data = aus_bowling_vs_sa 
        elif selected_team == "Australia" and opponent_team == "Sri Lanka":
            team_batting_data = aus_batting_vs_sri
            team_bowling_data = aus_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_aus
            opponent_bowling_data = sri_bowling_vs_aus
        elif selected_team == "Sri Lanka" and opponent_team == "Australia" :
            team_batting_data = sri_batting_vs_aus
            team_bowling_data = sri_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_sri
            opponent_bowling_data = aus_bowling_vs_sri    
        elif selected_team == "Australia" and opponent_team == "Afghanistan":
            team_batting_data = aus_batting_vs_afg
            team_bowling_data = aus_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_aus
            opponent_bowling_data = afg_bowling_vs_aus
        elif selected_team == "Afghanistan" and opponent_team == "Australia" :
            team_batting_data = afg_batting_vs_aus
            team_bowling_data = afg_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_afg
            opponent_bowling_data = aus_bowling_vs_afg     
        elif selected_team == "Australia" and opponent_team == "New Zealand":
            team_batting_data = aus_batting_vs_nz
            team_bowling_data = aus_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_aus
            opponent_bowling_data = nz_bowling_vs_aus
        elif selected_team == "New Zealand" and opponent_team == "Australia" :
            team_batting_data = nz_batting_vs_aus
            team_bowling_data = nz_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_nz
            opponent_bowling_data = aus_bowling_vs_nz
        elif selected_team == "Australia" and opponent_team == "West Indies":
            team_batting_data = aus_batting_vs_wi
            team_bowling_data = aus_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_aus
            opponent_bowling_data = wi_bowling_vs_aus
        elif selected_team == "West Indies" and opponent_team == "Australia" :
            team_batting_data = wi_batting_vs_aus
            team_bowling_data = wi_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_wi
            opponent_bowling_data = aus_bowling_vs_wi     
        elif selected_team == "Australia" and opponent_team == "Bangladesh":
            team_batting_data = aus_batting_vs_ban
            team_bowling_data = aus_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_aus
            opponent_bowling_data = ban_bowling_vs_aus
        elif selected_team == "Bangladesh" and opponent_team == "Australia" :
            team_batting_data = ban_batting_vs_aus
            team_bowling_data = ban_bowling_vs_aus
            opponent_batting_data = aus_batting_vs_ban
            opponent_bowling_data = aus_bowling_vs_ban   
        elif selected_team == "England" and opponent_team == "Pakistan":
            team_batting_data = eng_batting_vs_pak
            team_bowling_data = eng_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_eng
            opponent_bowling_data = pak_bowling_vs_eng
        elif selected_team == "Pakistan" and opponent_team == "England" :
            team_batting_data = pak_batting_vs_eng
            team_bowling_data = pak_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_pak
            opponent_bowling_data = eng_bowling_vs_pak    
        elif selected_team == "England" and opponent_team == "South Africa":
            team_batting_data = eng_batting_vs_sa
            team_bowling_data = eng_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_eng
            opponent_bowling_data = sa_bowling_vs_eng
        elif selected_team == "South Africa" and opponent_team == "England" :
            team_batting_data = sa_batting_vs_eng
            team_bowling_data = sa_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_sa
            opponent_bowling_data = eng_bowling_vs_sa      
        elif selected_team == "England" and opponent_team == "Sri Lanka":
            team_batting_data = eng_batting_vs_sri
            team_bowling_data = eng_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_eng
            opponent_bowling_data = sri_bowling_vs_eng
        elif selected_team == "Sri Lanka" and opponent_team == "England" :
            team_batting_data = sri_batting_vs_eng
            team_bowling_data = sri_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_sri
            opponent_bowling_data = eng_bowling_vs_sri       
        elif selected_team == "England" and opponent_team == "Afghanistan":
            team_batting_data = eng_batting_vs_afg
            team_bowling_data = eng_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_eng
            opponent_bowling_data = afg_bowling_vs_eng
        elif selected_team == "Afghanistan" and opponent_team == "England" :
            team_batting_data = afg_batting_vs_eng
            team_bowling_data = afg_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_afg
            opponent_bowling_data = eng_bowling_vs_afg            
        elif selected_team == "England" and opponent_team == "New Zealand":
            team_batting_data = eng_batting_vs_nz
            team_bowling_data = eng_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_eng
            opponent_bowling_data = nz_bowling_vs_eng
        elif selected_team == "New Zealand" and opponent_team == "England" :
            team_batting_data = nz_batting_vs_eng
            team_bowling_data = nz_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_nz
            opponent_bowling_data = eng_bowling_vs_nz    
        elif selected_team == "England" and opponent_team == "West Indies":
            team_batting_data = eng_batting_vs_wi
            team_bowling_data = eng_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_eng
            opponent_bowling_data = wi_bowling_vs_eng
        elif selected_team == "West Indies" and opponent_team == "England" :
            team_batting_data = wi_batting_vs_eng
            team_bowling_data = wi_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_wi
            opponent_bowling_data = eng_bowling_vs_wi     
        elif selected_team == "England" and opponent_team == "Bangladesh":
            team_batting_data = eng_batting_vs_ban
            team_bowling_data = eng_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_eng
            opponent_bowling_data = ban_bowling_vs_eng
        elif selected_team == "Bangladesh" and opponent_team == "England" :
            team_batting_data = ban_batting_vs_eng
            team_bowling_data = ban_bowling_vs_eng
            opponent_batting_data = eng_batting_vs_ban
            opponent_bowling_data = eng_bowling_vs_ban  
        elif selected_team == "Pakistan" and opponent_team == "South Africa":
            team_batting_data = pak_batting_vs_sa
            team_bowling_data = pak_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_pak
            opponent_bowling_data = sa_bowling_vs_pak
        elif selected_team == "South Africa" and opponent_team == "Pakistan" :
            team_batting_data = sa_batting_vs_pak
            team_bowling_data = sa_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_sa
            opponent_bowling_data = pak_bowling_vs_sa  
        elif selected_team == "Pakistan" and opponent_team == "Sri Lanka":
            team_batting_data = pak_batting_vs_sri
            team_bowling_data = pak_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_pak
            opponent_bowling_data = sri_bowling_vs_pak
        elif selected_team == "Sri Lanka" and opponent_team == "Pakistan" :
            team_batting_data = sri_batting_vs_pak
            team_bowling_data = sri_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_sri
            opponent_bowling_data = pak_bowling_vs_sri   
        elif selected_team == "Pakistan" and opponent_team == "Afghanistan":
            team_batting_data = pak_batting_vs_afg
            team_bowling_data = pak_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_pak
            opponent_bowling_data = afg_bowling_vs_pak
        elif selected_team == "Afghanistan" and opponent_team == "Pakistan" :
            team_batting_data = afg_batting_vs_pak
            team_bowling_data = afg_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_afg
            opponent_bowling_data = pak_bowling_vs_afg  
        elif selected_team == "Pakistan" and opponent_team == "New Zealand":
            team_batting_data = pak_batting_vs_nz
            team_bowling_data = pak_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_pak
            opponent_bowling_data = nz_bowling_vs_pak
        elif selected_team == "New Zealand" and opponent_team == "Pakistan" :
            team_batting_data = nz_batting_vs_pak
            team_bowling_data = nz_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_nz
            opponent_bowling_data = pak_bowling_vs_nz   
        elif selected_team == "Pakistan" and opponent_team == "West Indies":
            team_batting_data = pak_batting_vs_wi
            team_bowling_data = pak_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_pak
            opponent_bowling_data = wi_bowling_vs_pak
        elif selected_team == "West Indies" and opponent_team == "Pakistan" :
            team_batting_data = wi_batting_vs_pak
            team_bowling_data = wi_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_wi
            opponent_bowling_data = pak_bowling_vs_wi    
        elif selected_team == "Pakistan" and opponent_team == "Bangladesh":
            team_batting_data = pak_batting_vs_ban
            team_bowling_data = pak_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_pak
            opponent_bowling_data = ban_bowling_vs_pak
        elif selected_team == "Bangladesh" and opponent_team == "Pakistan" :
            team_batting_data = ban_batting_vs_pak
            team_bowling_data = ban_bowling_vs_pak
            opponent_batting_data = pak_batting_vs_ban
            opponent_bowling_data = pak_bowling_vs_ban   
       
        elif selected_team == "South Africa" and opponent_team == "Afghanistan":
            team_batting_data = sa_batting_vs_afg
            team_bowling_data = sa_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_sa
            opponent_bowling_data = afg_bowling_vs_sa
        elif selected_team == "Afghanistan" and opponent_team == "South Africa" :
            team_batting_data = afg_batting_vs_sa
            team_bowling_data = afg_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_afg
            opponent_bowling_data = sa_bowling_vs_afg  
        elif selected_team == "South Africa" and opponent_team == "New Zealand":
            team_batting_data = sa_batting_vs_nz
            team_bowling_data = sa_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_sa
            opponent_bowling_data = nz_bowling_vs_sa
        elif selected_team == "New Zealand" and opponent_team == "South Africa" :
            team_batting_data = nz_batting_vs_sa
            team_bowling_data = nz_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_nz
            opponent_bowling_data = sa_bowling_vs_nz      
        elif selected_team == "South Africa" and opponent_team == "West Indies":
            team_batting_data = sa_batting_vs_wi
            team_bowling_data = sa_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_sa
            opponent_bowling_data = wi_bowling_vs_sa
        elif selected_team == "West Indies" and opponent_team == "South Africa" :
            team_batting_data = wi_batting_vs_sa
            team_bowling_data = wi_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_wi
            opponent_bowling_data = sa_bowling_vs_wi    
        elif selected_team == "South Africa" and opponent_team == "Bangladesh":
            team_batting_data = sa_batting_vs_ban
            team_bowling_data = sa_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_sa
            opponent_bowling_data = ban_bowling_vs_sa
        elif selected_team == "Bangladesh" and opponent_team == "South Africa" :
            team_batting_data = ban_batting_vs_sa
            team_bowling_data = ban_bowling_vs_sa
            opponent_batting_data = sa_batting_vs_ban
            opponent_bowling_data = sa_bowling_vs_ban     
        elif selected_team == "Sri Lanka" and opponent_team == "Afghanistan":
            team_batting_data = sri_batting_vs_afg
            team_bowling_data = sri_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_sri
            opponent_bowling_data = afg_bowling_vs_sri
        elif selected_team == "Afghanistan" and opponent_team == "Sri Lanka" :
            team_batting_data = afg_batting_vs_sri
            team_bowling_data = afg_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_afg
            opponent_bowling_data = sri_bowling_vs_afg  
        elif selected_team == "Sri Lanka" and opponent_team == "New Zealand":
            team_batting_data = sri_batting_vs_nz
            team_bowling_data = sri_bowling_vs_nz
            opponent_batting_data = nz_batting_vs_sri
            opponent_bowling_data = nz_bowling_vs_sri
        elif selected_team == "New Zealand" and opponent_team == "Sri Lanka" :
            team_batting_data = nz_batting_vs_sri
            team_bowling_data = nz_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_nz
            opponent_bowling_data = sri_bowling_vs_nz     
        elif selected_team == "Sri Lanka" and opponent_team == "West Indies":
            team_batting_data = sri_batting_vs_wi
            team_bowling_data = sri_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_sri
            opponent_bowling_data = wi_bowling_vs_sri
        elif selected_team == "West Indies" and opponent_team == "Sri Lanka" :
            team_batting_data = wi_batting_vs_sri
            team_bowling_data = wi_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_wi
            opponent_bowling_data = sri_bowling_vs_wi   
        elif selected_team == "Sri Lanka" and opponent_team == "Bangladesh":
            team_batting_data = sri_batting_vs_ban
            team_bowling_data = sri_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_sri
            opponent_bowling_data = ban_bowling_vs_sri
        elif selected_team == "Bangladesh" and opponent_team == "Sri Lanka" :
            team_batting_data = ban_batting_vs_sri
            team_bowling_data = ban_bowling_vs_sri
            opponent_batting_data = sri_batting_vs_ban
            opponent_bowling_data = sri_bowling_vs_ban    
        elif selected_team == "New Zealand" and opponent_team == "West Indies":
            team_batting_data = nz_batting_vs_wi
            team_bowling_data = nz_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_nz
            opponent_bowling_data = wi_bowling_vs_nz
        elif selected_team == "West Indies" and opponent_team == "New Zealand" :
            team_batting_data = wi_batting_vs_nz
            team_bowling_data = wi_bowling_vs_nz
            opponent_batting_data = wi_batting_vs_nz
            opponent_bowling_data = wi_bowling_vs_nz     
        elif selected_team == "New Zealand" and opponent_team == "Bangladesh":
            team_batting_data = nz_batting_vs_ban
            team_bowling_data = nz_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_nz
            opponent_bowling_data = ban_bowling_vs_nz
        elif selected_team == "Bangladesh" and opponent_team == "New Zealand" :
            team_batting_data = ban_batting_vs_nz
            team_bowling_data = ban_bowling_vs_nz
            opponent_batting_data = ban_batting_vs_nz
            opponent_bowling_data = ban_bowling_vs_nz     
        elif selected_team == "West Indies" and opponent_team == "Bangladesh":
            team_batting_data = wi_batting_vs_ban
            team_bowling_data = wi_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_wi
            opponent_bowling_data = ban_bowling_vs_wi
        elif selected_team == "Bangladesh" and opponent_team == "West Indies" :
            team_batting_data = ban_batting_vs_wi
            team_bowling_data = ban_bowling_vs_wi
            opponent_batting_data = ban_batting_vs_wi
            opponent_bowling_data = ban_bowling_vs_wi  
        elif selected_team == "Afghanistan" and opponent_team == "West Indies":
            team_batting_data = afg_batting_vs_wi
            team_bowling_data = afg_bowling_vs_wi
            opponent_batting_data = wi_batting_vs_afg
            opponent_bowling_data = wi_bowling_vs_afg
        elif selected_team == "West Indies" and opponent_team == "Afghanistan" :
            team_batting_data = wi_batting_vs_afg
            team_bowling_data = wi_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_wi
            opponent_bowling_data = afg_bowling_vs_wi     
        elif selected_team == "Afghanistan" and opponent_team == "Bangladesh":
            team_batting_data = afg_batting_vs_ban
            team_bowling_data = afg_bowling_vs_ban
            opponent_batting_data = ban_batting_vs_afg
            opponent_bowling_data = ban_bowling_vs_afg
        elif selected_team == "Bangladesh" and opponent_team == "Afghanistan" :
            team_batting_data = ban_batting_vs_afg
            team_bowling_data = ban_bowling_vs_afg
            opponent_batting_data = afg_batting_vs_ban
            opponent_bowling_data = afg_bowling_vs_ban                    
                                                                       
                                                                                                                         
                                                                                  
                                      
        else:
            self.team_output.delete(1.0, tk.END)
            self.team_output.insert(tk.END, "Invalid team or opponent selection.")
            return
        
  
best_11 = select_best_11(team_batting_data, team_bowling_data, stadium_info)
        
        win_probability = calculate_winning_probability(team_batting_data, opponent_batting_data, team_bowling_data, opponent_bowling_data)

        self.team_output.delete(1.0, tk.END)

        for _, row in best_11.iterrows():
            player_type = "All-Rounder"
            if row['Player'] in team_batting_data['Player'].values and row['Player'] not in team_bowling_data['Player'].values:
                player_type = "Batsman"
            elif row['Player'] in team_bowling_data['Player'].values and row['Player'] not in team_batting_data['Player'].values:
                player_type = "Bowler"

            player_info = f"Player: {row['Player']} - Type: {player_type}"
            if player_type == "Batsman":
                player_info += f", Strike Rate: {row['strike_rate']:.2f}, Batting Avg: {row['avg']:.2f}"
            elif player_type == "Bowler":
                player_info += f", Bowling Avg: {row['avg']:.2f}, Economy: {row['econ']:.2f}"
            else:  # All-Rounder
                if 'strike_rate' in row and not pd.isna(row['strike_rate']):
                    player_info += f", Strike Rate: {row['strike_rate']:.2f}"
                if 'avg' in row and not pd.isna(row['avg']):
                    player_info += f", Batting Avg: {row['avg']:.2f}"
                if 'avg' in row and not pd.isna(row['avg']):
                    player_info += f", Bowling Avg: {row['avg']:.2f}"
                if 'econ' in row and not pd.isna(row['econ']):
                    player_info += f", Economy: {row['econ']:.2f}"

            self.team_output.insert(tk.END, f"{player_info}\n")
        
        # Show win probability
        win_probability_percentage = win_probability * 100
        self.team_output.insert(tk.END, f"\nWinning Probability: {win_probability_percentage:.2f}%")

root = tk.Tk()
app = CricketTeamSelector(root)
root.mainloop()
```
